```r
library(tidyverse)
```

```r
tree_data <- read_csv("Sx.htdata (1).csv")
tree_data
```

```r
# the impact of mean annual temperature
ggplot(tree_data, aes(x = P_MAT, y = Height, color = SITE)) +　
geom_point() + 
xlab("Mean temperature(province)") + 
ylab("Tree Height") + 
ggtitle("tree height v.s. mean temperature") +
theme(plot.title = element_text(color = "tomato", size = 19, face = "bold"))
```

```r
# the impact of mean coldest month temperature
ggplot(tree_data, aes(x = P_MCMT, y = Height, color = SITE)) +　
geom_point() + 
xlab("mean coldest month temperature(province)") + 
ylab("tree height") + 
ggtitle("tree height v.s. mean coldest month temperature") +
theme(plot.title = element_text(color = "tomato", size = 19, face = "bold"))
```

```r
# the impact of mean annual precipitation(rain)
ggplot(tree_data, aes(x = P_MAP, y = Height, color = SITE)) +　
geom_point() + 
xlab("mean annual percipitation(province)") + 
ylab("tree height") + 
ggtitle("tree height v.s. mean annual percipitation") +
theme(plot.title = element_text(color = "tomato", size = 19, face = "bold"))
```

```r
# the impact of mean summer precipitation(rain)
ggplot(tree_data, aes(x = P_MSP, y = Height, color = SITE)) +　
geom_point() + 
xlab("mean summer percipitation(province)") + 
ylab("tree height") +
ggtitle("tree height v.s. mean summer percipitation") +
theme(plot.title = element_text(color = "tomato", size = 19, face = "bold"))
```

```r
# the impact of latitude
ggplot(tree_data, aes(x = P_Latitude, y = Height, color = SITE)) +　
geom_point() + 
xlab("latitude(province)") + 
ylab("tree height") +
ggtitle("tree height v.s. latitude") +
theme(plot.title = element_text(color = "tomato", size = 19, face = "bold"))
```

```r
bxp <- ggplot(tree_data, aes(x = SITE, y = Height))+ 
  geom_boxplot(aes(fill = SITE)) +
  scale_fill_manual(values = c("#00AFBB", "#E7B800", "#FC4E07"))
bxp +
  theme(legend.position = "right")+ 
ggtitle("Height v.s. Site") +
theme(plot.title = element_text(color = "orangered", size = 19, face = "bold"))
```
